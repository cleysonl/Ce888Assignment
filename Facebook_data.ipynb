```python
facebook.dropna(0,inplace=True)
```

```python
lb = LabelBinarizer()
# Label Binarizer for Category (Action, Product and Inspiration)
lb_category=lb.fit_transform(facebook['Category'])
# Label Binarizer for Category (Link, Photo, Status and Video)
lb_type = lb.fit_transform(facebook['Type'])
lb_month = lb.fit_transform(facebook['Post Month'])
lb_weekday = lb.fit_transform(facebook['Post Weekday'])
lb_hour = lb.fit_transform(facebook['Post Hour'])
```

```python
# Label Binarizer for Category (Action, Product and Inspiration)
Action =[]
Product =[]
Inspiration=[]
Link=[]
Photo=[]
Status=[]
Video=[]
for i in range(len(facebook)):
    Action.append(lb_category[i][0])
    Product.append(lb_category[i][1])
    Inspiration.append(lb_category[i][2])
    Link.append(lb_type[i][0])
    Photo.append(lb_type[i][1])
    Status.append(lb_type[i][2])
    Video.append(lb_type[i][3])
    
facebook['Action']=Action
facebook['Product']=Product
facebook['Inspiration']=Inspiration
facebook['Link']=Link
facebook['Photo']=Photo
facebook['Status']=Status
facebook['Video']=Video
bin_var = [lb_month,lb_weekday,lb_hour]
for i in bin_var:
    i=pd.DataFrame(i)
    facebook=pd.concat([facebook,i])
facebook.columns
```
